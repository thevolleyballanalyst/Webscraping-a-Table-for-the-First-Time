```python
from bs4 import BeautifulSoup #helps in web scraping
import requests # helps us to download a web page
```

```python
import pandas as pd # helps us download pd for Data Frames
```

```python
url ='https://thevolleyballanalyst.wordpress.com/2021/03/29/establishing-weights-for-the-quality-of-the-pass-addressing-the-weight-of-reception-errors-and-overpasses/'

#defining the web address as url
```

```python
data = requests.get(url).text

#convert the website into text
```

```python
soup = BeautifulSoup(data,"html5lib")

# it parses(?) the document.  not sure what the html5lib is 
```

```python
tables = soup.find_all('table')

# finds all tables on the webpage
```
